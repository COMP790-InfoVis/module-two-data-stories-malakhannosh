<h1>Nobel Prize Winners</h1>

The dataset used was sourced from [Kaggle](https://www.kaggle.com/datasets/bahramjannesarr/nobel-prize-from-1901-till-2020).

```python
import pandas as pd

data= pd.read_csv("nobel_final.csv")
data.head()
```

```python
import altair as alt

prizes = data['year'].value_counts().reset_index()
prizes.columns = ['year', 'count']

alt.Chart(prizes).mark_line().encode(
    x=alt.X('year:N', axis=alt.Axis(title='Year')),
    y=alt.Y('count:Q', axis=alt.Axis(title="Number of Winners"))
).properties(
    title='Number of Nobel Prize Winners Each Year',
    width=700
).interactive()
```
